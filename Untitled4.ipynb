```python
import pandas as pd

# URL of the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

# Read the dataset
data = pd.read_csv(url, delimiter=';')

# Save it as a CSV file locally
data.to_csv("winequality-red.csv", index=False)
print("File saved as 'winequality-red.csv'")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load dataset
data = pd.read_csv('winequality-red.csv')

# Add binary quality labels
data['quality_label'] = np.where(data['quality'] >= 7, 1, 0)

# Features and target
X = data.drop(columns=['quality', 'quality_label'])
y = data['quality_label']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Train and evaluate SVM with different kernels
kernels = ['linear', 'poly', 'rbf']
results = {}

# Adjust class imbalance with class weights
class_weights = dict(enumerate(
    compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
))

for kernel in kernels:
    print(f"Evaluating kernel: {kernel}")

    # Initialize the model with class weights
    model = SVC(kernel=kernel, class_weight=class_weights, random_state=42)
    model.fit(X_train, y_train)

    # Predict on test set
    y_pred = model.predict(X_test)

    # Compute accuracy
    accuracy = accuracy_score(y_test, y_pred)
    results[kernel] = accuracy

    # Print classification report with zero_division to avoid warnings
    print(classification_report(y_test, y_pred, zero_division=0))

# Plot kernel performance
plt.bar(results.keys(), results.values(), color=['blue', 'orange', 'green'])
plt.title("SVM Kernel Performance")
plt.ylabel("Accuracy")
plt.xlabel("Kernel")
plt.show()
```

```python
from sklearn.decomposition import PCA

# Reduce features to 2D for visualization
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Train SVM with RBF kernel on reduced data
model_rbf = SVC(kernel='rbf', random_state=42)
model_rbf.fit(X_train_pca, y_train)

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(X_train_pca[:, 0].min(), X_train_pca[:, 0].max(), 100),
                     np.linspace(X_train_pca[:, 1].min(), X_train_pca[:, 1].max(), 100))
Z = model_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, edgecolor='k')
plt.title("SVM Decision Boundary (RBF Kernel)")
plt.show()
```
